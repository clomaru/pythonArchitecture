## GILとは

> プログラミング言語のインタプリタのスレッドによって保持されるスレッドセーフでないコードを、他のスレッドと共有してしまうことを防ぐための排他ロックである。  


PythonにはGILがあるので、Pythonのバイトコードを一度に1スレッドしか行えない.    
たとえ、マルチスレッドになっていたとしても、実際には並行処理を擬似的に実現するために、インタプリタが交互にロックを解放・獲得しているに過ぎない。  


### CPUバウンドの場合

pythonインタープリタは内部でticksというカウンタを持っている  
スクリプト実行時に一定期間のticks(コードの塊)でスレッドの実行がスイッチされる  


### I/Oバウンドの場合

ブロッキングI/Oを多く発行するようなスクリプトの場合、I/O要求を出す前に、スレッドがGILを解放  
I/Oから戻ったときにGILを取得する

## GILの影響
PythonはGILがあるので、マルチスレッドにしても同時に実行されるスレッドは1つに制限されている.  
しかし、I/O待ちの間や、特定のサードパーティの拡張機能により、Cのコードを並列実行できたりする。 

なので、CPythonはマルチスレッドをサポートしているにも関わらず、GILがあるので同時に１つのスレッドしか進行できないようになっているので、結局並列計算できていない。むしろ、スレッド立ち上げなどにコストがかかる。

- Pythonがある種のシステムコールを行う際に生じるブロッキングI/Oを扱うため

__★★Pythonはシステムコールを行うときは、直前にGILを解放し、終わるとGILを獲得するようになっているので、システムコールの際はGILの制約を受けることなく並列処理を実行することができる★★__

 ## multiprocessingの使い方
 
 ### Processを使う
 

```python
import time
import multiprocessing

def worker(data):
    [calc(x) for x in data]

def calc(x):
    a = x * 2
    print(a)
    time.sleep(2)

if __name__ == '__main__':
    split_data = [[1,2,3],[4,5]]

    jobs = []
    for data in split_data:
        job = multiprocessing.Process(
            target=worker, # target:並列実行する関数
            args=(data,) # target関数の引数. タプル
        )
        jobs.append(job)
        job.start()

    # join()メソッドが呼ばれたプロセスは処理が終了するまでブロックする。
    # つまりこの場合、join()がないと先に'finish'が出力される
    [job.join() for job in jobs]

    print('finish')


# 参考
# https://librabuch.jp/blog/2015/01/python-multiprocessing/
```

### submit
- fn関数をリソースプールで実行するようにスケジュールし、Futureオブジェクトを返す。  
- １つのタスクを実行キューに追加する

- 戻り値のFutureオブジェクトでタスクのキャンセルや実行結果の取得を行う

### map

- 戻り値はタスクの実行結果を取得するためのgenerator

### shutdown

## ThreadPoolExecutor
スレッドプールを表す




- スレッドを使って並列タスクを実行する
- ネットワークアクセスなどCPUに負荷がかからない処理の並列実行に適している

workerのデフォルトはos.cpu_count()の5倍の数字。  


effective python p.146「スレッドプールの開始や通信のオーバーヘッドのために、かえって遅くなりました」  



### スレッドプールとは
スレッドを管理する効率的なアプローチを提供する。  
スレッドは生成や切り替えにコストがかかる。なので、一度使ったスレッドを可能な限り使いまわすなど
- 一定数のスレッドを常に立ておく
- タスクを待たせておくためのキューを持つ
- 現在処理中のタスクが終わり次第、次のタスクをキューから取り出して実行する　　
- http://ufcpp.net/study/csharp/misc_task.html

## ProcessPoolExecutor
プロセスプールを表す
- プロセスを使って並列タスクを実行する
- CPUに負荷がかかる計算処理などの並列実行に適している

引数のmax_workerはデフォルではos.cpu_count()が返すCPU数になっている  

このppeの真価は、CPU不可の高いジョブで発揮される。  



### 向いているタスク
「分離した、レバレッジの高いタスクに向いていてる」  
逆にこのような特性がない場合は、multiproseccingのオーバーヘッドは、並列性による速度向上をもたらさない。
#### 分離
→プログラムの他の部分と状態を共有する必要がない
#### レバレッジが高い
少量のデータだけを親と子のプロセス間でやりとりするだけで、大量の計算ができるということ




### 注意点
https://qiita.com/kokumura/items/2e3afc1034d5aa7c6012
> -  ProcessPoolExecutorはプロセス間通信によって実現しているため、いくつか制限がある。
    - 関数の引数および戻り値は、pickleを使ってシリアライズ可能なオブジェクトでなければならない。
    - 関数自体もプロセス間で渡せなければならない。インスタンスメソッドはだめ。ラムダ式はOK。
    - 関数の中で副作用としてグローバル変数を書き換えるなどしても、呼び元のプロセスには反映されない。

fluent python p.542
> 国旗ダウンロードの例やI/Oの制約のあるジョブでは、ProcessPoolExecutorを使用するメリットはありません。

## 協調的マルチタスク
expert p.459  
OSのコンテキストスイッチ（スレッドやプロセスを切り替えるやつ）に頼ることをしない。  
代わりに、各プロセスが待機状態になったら自発的に制御を解放して、同時に実行されている多数の他の処理に制御権を渡す。  

アプリケーションレベルでの協調的マルチタスクを行うときは、複数のスレッドやプロセスを強調させるのではなく、全てを１つのスレッドの中で実行する。  

### ノンプリエンプティブである
ここが、マルチスレッドを使うのと異なる点である。  
マルチスレッドの場合は、pythonのスレッドがシステムレベルのスレッドとして実装されているので、OSは現在動いているスレッドにいつでも割り込んで、他のスレッドに制御を渡せる。  
一方、非同期プログラミングでは、タスクはイベントループによって割り込まれない。なので、協調的マルチタスクはノンプリエンプティブマルチタスクとも呼ばれる。  

厳密にはシステムスケジューラの割り込みで処理を一時中断するが、制御が戻されたときは、まさにその場所から開始する（ジェネレータっぽい）  


## Futureオブジェクトとは
http://iuk.hateblo.jp/entry/2017/01/27/173449
>　Futureオブジェクトは、簡単にいえば __結果の代わりとなる__ オブジェクトです。ある値を渡すような処理があったとして、送信側はとりあえずFutureオブジェクトを受信側に渡し、送信側は処理が終わり次第そのFutureオブジェクトに値を放り込みます。受信側は、Futureオブジェクトの状態が完了になり次第、値を取り出して続きの処理に使うことができます。

## asyncとawait

### async
asyncはdef文の前に使用し、新しいコルーチンを定義する。  
コルーチン関数の実行は、厳密に定義された状況に応じて中断、再開される。   

これによって定義した関数を呼び出しても、その場では関数内のコードを実行せず、コルーチンオブジェクトを返す。  
コルーチンオブジェクトはイベントループの中で順番が来るまで何もsに愛。

### await
コルーチンやFutureが結果を返すのを待ち、実行が終わるまで協調的に制御を解放してイベントループに渡す。
